### Data Visualization in Python

Data visualization is the graphical representation of data. It helps us understand patterns, trends, comparisons, and relationships in data.

## Libraries Used
### #️⃣ Matplotlib

A basic and powerful plotting library in Python

Gives full control over plots

```python
import matplotlib.pyplot as plt
```

## 2️⃣ Seaborn
Built on top of Matplotlib

Provides better-looking and statistical plots with less code

```python
import seaborn as sns
```

# pandas for data set loading

```python
import pandas as pd

#Data set loading
df=pd.read_csv(r"F:\Downloads\large_students.csv")
```

### Bar chart
plt.bar() → creates a bar chart

```python
#MATPLOT
plt.bar(df['Marks'],df['Age'])
plt.title("Bar Chart of Marks")
plt.xlabel('Marks')
plt.ylabel('AGe')
plt.show()
```

```python
#SEABORN
sns.barplot(x='Marks',y='Age',data=df)
plt.show
```

### Line chart
Shows trends over continuous data (e.g., change over time)
plt.plot() → creates a line plot

xlabel() → sets x-axis label

ylabel() → sets y-axis label

title() → adds a title

show() → displays the plot

```python
# MATPLOT
plt.plot(df['Name'], df['Marks'])
plt.xlabel("Students")          
plt.ylabel("Marks")             
plt.title("Line Plot of Marks") 
plt.show()   
```

```python
#SEABORN
#automatically styles
sns.lineplot(x='Name',y='Marks',data=df)
plt.title('name and marks of students')
plt.show()
```

### histogram
show data distribution and frequency

bins-interval

kde-gives smooth curve

```python
#MATPLOT
plt.hist(df['Marks'], bins=5)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram of Marks")
plt.show()
```

```python
#SEABORN
sns.histplot(df['Marks'] ,kde=True , bins=5)
plt.title("students marks")
plt.show()
```

### Scatter plot
Shows relationship between two numerical variables

```python
#MATPLOT
plt.scatter(df['Age'], df['Marks'])
plt.xlabel("Age")
plt.ylabel("Marks")
plt.title("Scatter Plot: Age vs Marks")
plt.show()
```

```python
#SEABORN
sns.scatterplot(x='Marks',y='Age',data=df)
plt.title("age and marks")
plt.show()
```
