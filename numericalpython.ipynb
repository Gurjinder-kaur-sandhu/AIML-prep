# ARRAY CREATION
### array()

array() is the main function of NumPy, which converts a list or tuple into a NumPy array.

An array is a multidimensional container that allows you to store and operate on numbers efficiently.


```python
import numpy as np
a=np.array([1,2,3,4]) #1d array
print(a)

b=np.array([[1,2],[2,4]])
print(b)
```

### shape()
shape tells the dimensions of the array.

For 2D arrays, it is given as a tuple (rows, columns)

```python
print(a.shape)
print(b.shape)
```

### reshape()

reshape() allows you to change the size or dimensions of an array without changing its elements.

The total number of elements must remain the same.

```python
c=np.array([1,2,3,4,4,3])
c2=c.reshape(2,3)
print(c2)
```

### Basic Operations: + , - , * , /

With NumPy arrays, mathematical operations are element-wise.

That means the operation is applied to each element individually.

```python
x=np.array([7,4,3,2,2])
y=np.array([6,8,4,4,3])
z=x+y
print(z)
print(x*y)
print(x/y)
print(x-y)
```

### Statistical Operations: mean(), max(), min()

mean() → Calculates the average of array elements

max() → Finds the largest element

min() → Finds the smallest element

```python
print(x.mean())
print(x.max())
print(x.min())
print(b.mean(axis=0))  # Column-wise mean
print(b.mean(axis=1))  # Row-wise mean 

```
