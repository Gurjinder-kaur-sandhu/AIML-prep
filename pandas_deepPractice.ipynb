### Fill missing values (mean)

Description: Replace missing values with the column’s average value.

```python
df=pd.read_csv(r"F:\Downloads\large_students.csv")

df['Marks']=df['Marks'].fillna(df['Marks'].mean())
print(df)
```

### Fill missing values (fixed value)

Description: Replace missing values with a fixed value like 0.

```python
df['Age']=df['Age'].fillna(0)
```

### Drop rows with missing data

Description: Remove rows that contain any missing values.

```python
df_1=df.dropna()
```

### Sort values

Description: Sort the dataset based on a specific column.

```python
df.sort_values(by="Marks" ,ascending=False)
```
