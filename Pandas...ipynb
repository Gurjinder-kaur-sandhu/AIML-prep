# Pandas Introduction

Pandas is a Python library used for data manipulation and analysis.

It provides DataFrames (like Excel tables) and Series (like columns).

You can read data, clean it, analyze it, and filter it easily.

### read_csv()

read_csv() is used to read CSV (Comma Separated Values) files into a Pandas DataFrame

```python
import pandas as pd
df=pd.read_csv("large_students.csv")
```

### head()

Shows the first 5 rows of the DataFrame by default.

Useful to quickly check data.

```python
df.head()
#df.head(10)
```

### tail()

Shows the last 5 rows by default.

```python
df.tail()
```

### info()

Gives summary of DataFrame:

Number of rows & columns

Column names

Data types

Non-null counts

```python
df.info()
```

### Column Selection

Columns in a DataFrame can be selected using brackets [] or dot notation.

col1 is a Series (single column)

cols is a DataFrame (multiple columns)

```python
# Select single column
col1 = df['Name']

# Select multiple columns
col2 = df[['Name', 'Marks']]
print(col1,col2)
```

### Row Filtering (using loc)

loc is used to select rows and columns by labels.

Syntax: df.loc[row_condition, column_selection]

Examples:

```python
df_fil1=df.loc[df['Age']>10]
df_fil=df.loc[df['Age']>10,['Name','Marks']]
print(df_fil1,df_fil)
```

### Mean of Marks:

```python
mean_marks=df['Marks'].mean()
print(mean_marks)
```

### max of marks

```python
max_marks=df['Marks'].max()
print(max_marks)
```

### plotting of csv file

```python
import matplotlib.pyplot as plt
plt.bar(df['Marks'],df['Age'])
plt.xlabel("Marks")
plt.ylabel("Age")
plt.title("students marks")
plt.show()

```
