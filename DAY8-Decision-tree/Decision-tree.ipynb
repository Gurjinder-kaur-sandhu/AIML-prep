# load dataset

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
```

### DATA KEYS

```python
print(data.keys())
```

### output
malignant-0 (cancer +ve)
benign -1(cancer -ve)

```python
print(data.target_names)
```

```python
X=data.data
Y=data.target
```

### test train splitting 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y , test_size=0.2,random_state=42)
```

# Create Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(
    criterion="gini",     
    random_state=42
)
```

DecisionTreeClassifier → classification model

criterion="gini" → split based on Gini impurity

### training of model fit ()

```python
# train the model
model.fit(X_train,Y_train)
```

### prediction

```python
# prediction
y_pre=model.predict(X_test)
```

### ACCURACY

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_test, y_pre)
print("Accuracy:", accuracy)
```

### CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, y_pre)
print(cm)
```

### PRINTING OUTPUT OF PREDICTIONS

```python
print(y_pre[:10])
```

# OVERFITTING AND UNDERFITIING

Controlling Overfitting using max_depth

max_depth limits how deep the decision tree can grow.

Smaller depth → simpler model

Larger depth → more complex model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model_depth_2 = DecisionTreeClassifier(max_depth=2, random_state=42)
model_depth_2.fit(X_train, Y_train)

y_pred_2 = model_depth_2.predict(X_test)

print("Accuracy with max_depth=2:", accuracy_score(Y_test, y_pred_2))
```

### max_depth=2

-Tree is very shallow

-Model may underfit

-Accuracy is usually moderate

```python
model_depth_5 = DecisionTreeClassifier(max_depth=5, random_state=42)
model_depth_5.fit(X_train, Y_train)

y_pred_5 = model_depth_5.predict(X_test)

print("Accuracy with max_depth=5:", accuracy_score(Y_test, y_pred_5))
```

### max_depth=5

-Tree is deeper but controlled

-Learns important patterns

-Ignores noise

-Test accuracy is usually better
